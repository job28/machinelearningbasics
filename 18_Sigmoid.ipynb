```python
import math
import matplotlib.pyplot as plt

x=[]
y=[]
for z in range(-5, 5):
    x.append(z)
    y.append(1/(1 + math.exp(-z)))
print(y)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid function")
plt.savefig('plots/sigmoid.png')
plt.show()
```
