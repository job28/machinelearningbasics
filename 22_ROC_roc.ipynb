```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data/rocExample.csv")
# print(rocData.actual)
df.head()
```

```python
X = df.drop(columns='actual')
y = df["actual"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=10, random_state=1, stratify=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
print('Accuracy', logmodel.score(X_test, y_test))
```

```python
predictions = logmodel.predict(X_test)
# print(predictions)
# print(y_test)
```

```python
cm = confusion_matrix(y_test, predictions)
print(cm)

tn, fp, fn, tp = cm.ravel()
print("TN", tn, "FP", fp, "FN", fn, "TP", tp)
```

```python
probs = logmodel.predict_proba(X_test)
#print(probs)
probs = probs[:,1]
print(probs)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt

fpr, tpr, _ = roc_curve(y_test, probs)
plt.plot(fpr, tpr)

plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
```
