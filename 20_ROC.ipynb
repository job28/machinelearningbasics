```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
rocData = pd.read_csv("data/rocExample.csv")
print(rocData.actual)
N = 10
P = 10
total = 20
fprList = list()
tprList = list()
tList = list()
```

```python
for i in range(total) :
    predicted = np.zeros(20)
    TPVector = np.zeros(20)
    FPVector = np.zeros(20)
 
    threshold = rocData.score[i]
    tList.append(threshold)
    for j in range(total) :
        if rocData.score[j] >= threshold :
            predicted[j] = 1
    for j in range(total) :
        if rocData.actual[j] == 'p' and predicted[j] == 1 :
            TPVector[j] = 1
        if rocData.actual[j] == 'n' and predicted[j] == 1 :
            FPVector[j] = 1
 
    TPR = sum(TPVector) / P
    FPR = sum(FPVector) / N

    
    
    fprList.append(FPR)
    tprList.append(TPR)
```

```python
plt.scatter(fprList, tprList, color='black')
plt.plot(fprList, tprList, color='black')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.savefig('plots/roc1.png')
plt.show()
# plt.figure()
fig, ax = plt.subplots()
for i, txt in enumerate(tList):
    ax.annotate(txt, (fprList[i], tprList[i]))
    plt.scatter(fprList[i], tprList[i], color='black')
plt.plot(fprList, tprList, color='black')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.savefig('plots/roc2.png')
plt.show()
```
