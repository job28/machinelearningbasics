```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([6,1,9,5,17,12])
print("x.shape ", x.shape)
```

```python
model = LinearRegression()
model.fit(x,y)
print(model.intercept_)
print(model.coef_)
```

```python
y_hat = model.predict(x)
print(y_hat)
```

```python
plt.scatter(x,y,color='black')
plt.plot(x,y_hat, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x,y,color='black', label='actual')
plt.scatter(x,y_hat,color='blue', label='predicted')
plt.plot(x,y_hat, color='blue', linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.title('Predicted and actual values of y')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

yhat = model.predict(X_test)
print(mean_squared_error(y_test, yhat, squared=False))
```
